```python
import pandas as pd
dataset = pd.read_csv('Titanic Dataset.csv' )
dataset
```

```python
#Encoding SEX Column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = ct.fit_transform(dataset)
X
```

```python
#Printing the new set
df=pd.DataFrame(X,columns=['female','male','PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare'])
df
```

# Spiliting and Scaling part

```python
#performing scaling to Fare column
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(X, test_size = 0.2, random_state=50)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 8:] = sc.fit_transform(X_train[:, 8:])
X_test[:, 8:] = sc.transform(X_test[:, 8:])
X_train
```
